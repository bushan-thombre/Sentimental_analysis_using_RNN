```python
df= pd.read_csv("D:/projects/Downloads/swiggy.csv")
df.head()
```

```python
df["Review"] = df["Review"].str.lower()
```

```python
df["Review"] = df["Review"].str.replace(r'[^a-z0-9\s]' ,'',regex = True)
```

## Tokenizer

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['Review'])
```

```python
tokenizer.word_index
```

```python
max_feature = 1000
max_length = 200
```

```python
tokenizer.word_counts
```

```python
tokenizer.texts_to_sequences(df['Review'])
```

## Padding

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
x = pad_sequences(tokenizer.texts_to_sequences(df['Review']),maxlen =max_length)
y = df['Sentiments']
```

```python
x
```

```python
y
```

## Split the datasets and train_test

```python
from sklearn.model_selection import train_test_split
xtrain,xtest ,ytrain ,ytest = train_test_split(x,y ,test_size = 0.2, random_state = 42)
```

## model building using rnn

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN ,Dense ,Embedding
```

```python
model = Sequential()
model.add(Embedding(input_dim = 1000 ,output_dim = 16 ,input_length = 200))
model.add(SimpleRNN(64 ,activation = 'tanh' ,return_sequences = False))
model.add(Dense(1 ,activation = 'sigmoid'))
model.build(input_shape = (None ,1952))
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy' ,optimizer = 'adam' ,metrics =['accuracy'])
```

```python
model.fit(xtrain,ytrain ,epochs = 5 ,validation_data =(xtest,ytest))
```

```python
def predict_sentiments(review_text):
    text =review_text.lower()
    text = re.sub(r'[^a-z0-9\s]','',text)

    seq = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(seq , maxlen = max_length)

    prediction = model.predict(padded)[0][0]
    return f"{'Positive' if prediction >= 0.5 else 'Negative'} (Probability: {prediction:.2f})"
```

```python
sample_review = "The food was great."
print(f"Review: {sample_review}")
print(f"Sentiment: {predict_sentiments(sample_review)}")
```
